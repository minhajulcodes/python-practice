```python

# -----------------------------
# 1 থেকে 10 প্রিন্ট করা
# -----------------------------
print("While loop দিয়ে 1 থেকে 10:")
i = 1
while i <= 10:
    print(i, end=" ")
    i += 1
print("\n")

print("For loop দিয়ে 1 থেকে 10:")
for i in range(1, 11):
    print(i, end=" ")
print("\n")

# -----------------------------
# 10 থেকে 1 প্রিন্ট করা
# -----------------------------
print("While loop দিয়ে 10 থেকে 1:")
i = 10
while i >= 1:
    print(i, end=" ")
    i -= 1
print("\n")

print("For loop দিয়ে 10 থেকে 1:")
for i in range(10, 0, -1):
    print(i, end=" ")
print("\n")

# -----------------------------
# 1 থেকে 10 এর জোড় সংখ্যা
# -----------------------------
print("While loop দিয়ে 1 থেকে 10 এর জোড় সংখ্যা:")
i = 2
while i <= 10:
    print(i, end=" ")
    i += 2
print("\n")

print("For loop দিয়ে 1 থেকে 10 এর জোড় সংখ্যা:")
for i in range(2, 11, 2):
    print(i, end=" ")
print("\n")

# -----------------------------
# 1 থেকে 10 এর বিজোড় সংখ্যা
# -----------------------------
print("While loop দিয়ে 1 থেকে 10 এর বিজোড় সংখ্যা:")
i = 1
while i <= 10:
    print(i, end=" ")
    i += 2
print("\n")

print("For loop দিয়ে 1 থেকে 10 এর বিজোড় সংখ্যা:")
for i in range(1, 11, 2):
    print(i, end=" ")
print("\n")
```
